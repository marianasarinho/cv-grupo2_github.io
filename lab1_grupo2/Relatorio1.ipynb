### PARTE 1: Processamento básico nas imagens e vídeos

**(A) Leitura de Imagem em Arquivo**

O programa original ("L__1_img.py") lê a imagem `messi5.jpg` e a exibe em uma janela do OpenCV. O ponto principal da atividade é entender por que a imagem é exibida em escala de cinza. Era realizada a leitura da imagem através do comando:

<div align="center">
img = cv.imread('messi5.jpg',0)
</div>

O segundo parâmetro igual a 0 indica que a imagem deve ser carregada em escala de cinza e, por esse motivo, a janela exibida pelo opencv apresenta a imagem sem cores. Para exibir a imagem colorida, o programa foi modificado para:

<div align="center">
img = cv.imread('messi5.jpg',1)
</div>

Nesse caso, o valor 1 faz com que a imagem seja carregada em cores, preservando os canais RBG presentes no arquivo.

Código original:

```python
# Bibliotecas necessárias
img = cv.imread('messi5.jpg',0)
cv.imshow('image',img)
k = cv.waitKey(0)
if k == 27:         # wait for ESC key to exit
    cv.destroyAllWindows()
elif k == ord('s'): # wait for 's' key to save and exit
    cv.imwrite('messigray.png',img)
    cv.destroyAllWindows()
```

Código corrigido:

```python
img = cv.imread('messi5.jpg',1)
cv.imshow('image',img)
k = cv.waitKey(0)
if k == 27:         # wait for ESC key to exit
    cv.destroyAllWindows()
elif k == ord('s'): # wait for 's' key to save and exit
    cv.imwrite('messigray.png',img)
    cv.destroyAllWindows()
```

**(B) Leitura de Vídeo em Arquivo**

O programa ("L__2_video.py") reproduzia um vídeo armazenado em arquivo utilizando a leitura sequencial dos quadros (frames). A velocidade de reprodução foi controlada pela instrução:

<div align="center">
time.sleep(1/25.0) # 25 fps
</div>

O comando determina o intervalo de tempo entre a exibição de quadros (ou *frames*) consecutivos: quanto menor o tempo de espera, maior será a velocidade de reprodução. Da mesma forma, quanto maior o tempo de espera, mais lenta será a exibição do vídeo. Para acelerar a reprodução, foi utilizado um valor correspondente a aproximadamente 250 quadros por segundo (fps): 

<div align="center">
time.sleep(1/250.0)
</div>

Com essa alteração, o atraso entre os frames foi reduzido singificativamente, aumentando a velocidade de exibição. 

Código modificado:

```python
cap = cv.VideoCapture('big_buck_bunny.mp4')

while(cap.isOpened()):
    ret, frame = cap.read()
    
    if ret==True:
        # show the frame
        cv.imshow('frame',frame)

        #wait next frame by 40ms - 25fps
        time.sleep(1/250.0) 
        
        if cv.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
               
cap.release()
cv.destroyAllWindows()
```

**(C) Leitura de Imagem em Câmera**

O programa ("L__3_webcam.py") realizava apenas a captura e exibição contínua dos quadros provenientes da webcam. A modificação feita permite salvar uma fotografia quando a tecla 'x' fosse pressionada pelo usuário, através da verificação da tecla pressionada durante a execução do programa.

Código original:

```python
cap = cv.VideoCapture(0)

if not cap.isOpened():
    print("Cannot open camera")
    exit()
    
while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    # if frame is read correctly ret is True
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
    
    # Display the resulting frame
    cv.imshow('frame', frame)
    
    if cv.waitKey(1) == ord('q'):
        break

# When everything done, release the capture
cap.release()
cv.destroyAllWindows()
```

Código alterado:

```python
cap = cv.VideoCapture(0)

if not cap.isOpened():
    print("Erro: Não foi possível abrir a câmera.")
    exit()

print("Pressione 'x' para tirar uma foto ou 'q' para sair.")

while True:
    ret, frame = cap.read()
    
    if not ret:
        print("Erro: Não foi possível receber o frame. Saindo...")
        break

    cv.imshow('frame', frame)

    # Armazena o valor da tecla pressionada
    key = cv.waitKey(1) & 0xFF

    # Se a tecla 'q' for pressionada, fecha o programa
    if key == ord('q'):
        break
        
    # Se a tecla 'x' for pressionada, salva a foto atual
    elif key == ord('x'):
        cv.imwrite('foto1.png', frame)
        print("Imagem salva com sucesso como 'foto1.png'!")

# Libera a câmera e fecha todas as janelas abertas
cap.release()
cv.destroyAllWindows()
```

A variável `key` armazena a tecla pressionada. Quando `key == ord('x')`, o frame atual da câmera é salvo por `cv.imwrite('foto1.png', frame)`. Assim, a imagem registrada corresponde ao quadro que estava sendo visualizado no instante em que a tecla `x` foi pressionada.

**(D) Gravação de Vídeo em Câmera**

O programa ("L__4_webcap.py") original realizava a captura dos quadros da webcam e armazenava o resultado em um arquivo AVI. Para que a imagem gravada fique normal, foi removida a inversão vertical do frame. Também foi escolhida uma taxa de quadros de `10 fps`, suficiente para uma gravação simples de webcam e compatível com os vídeos gerados pelo grupo.

Ponto adequado para processamento: caso seja necessário aplicar alguma operação de processamento, como conversão para escala de cinza, filtragem, recorte, espelhamento ou detecção de bordas, essa operação deve ser realizada logo após a leitura do frame (`ret, frame = cap.read()`) e antes de `cv.imshow`, `cv.imwrite` ou `out.write`. Isso garante que a imagem processada seja a mesma exibida ou salva.

Código modificado:

```python
cap = cv.VideoCapture(0)

if not cap.isOpened():
    raise RuntimeError('Não foi possível abrir a câmera')

width = int(cap.get(cv.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv.CAP_PROP_FRAME_HEIGHT))

# FPS adequado para uma gravação simples de webcam.
fps = 10.0

fourcc = cv.VideoWriter_fourcc(*'XVID')
out = cv.VideoWriter('saida.avi', fourcc, fps, (width, height))

print("Pressione 'q' para encerrar a gravação.")

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print('Não foi possível receber o frame. Encerrando...')
        break

    # Qualquer processamento da imagem deve ser feito aqui,
    # depois da leitura do frame e antes da exibição/gravação.

    out.write(frame)
    cv.imshow('frame', frame)

    if cv.waitKey(int(1000 / fps)) & 0xFF == ord('q'):
        break

cap.release()
out.release()
cv.destroyAllWindows()
```

Na Atividade C, a webcam foi usada como fonte de entrada em tempo real. A modificação principal consistiu em associar a tecla `x` ao comando `cv.imwrite`, permitindo salvar o frame atual como imagem estática.
